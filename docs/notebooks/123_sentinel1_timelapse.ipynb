```python
import ee
import geemap
import ffmpeg
import warnings
```

```python
Map = geemap.Map()
Map
```

```python
ee.Authenticate()
ee.Initialize(project='ee-pyjuliar')
```

Pan and zoom to an area of interest and draw a rectangle on the map.

```python
roi = Map.user_roi
if roi is None:
    roi = ee.Geometry.BBox(96.10, 20.54, 96.2, 20.6)
    Map.addLayer(roi)
    Map.centerObject(roi)
```

```python
timelapse = geemap.sentinel1_timelapse(
    roi,
    out_gif="sentinel1.gif",
    start_year=2024,
    end_year=2025,
    start_date="08-01",
    end_date="06-20",
    frequency="day",
    vis_params={"min": -30, "max": 0},
    palette="Greys",
    frames_per_second=1,
    title="Sentinel-1 Timelapse",
    add_colorbar=True,
    colorbar_bg_color="gray",
)
```

```python
geemap.show_image(timelapse)
```
